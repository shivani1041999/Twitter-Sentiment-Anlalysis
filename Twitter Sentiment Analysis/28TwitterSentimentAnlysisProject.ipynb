```python
import string
from nltk.corpus import stopwords
punctations = list(string.punctuation)
stop = stopwords.words('english')
stop = punctations + stop
# stop
```

```python
from nltk.tokenize import word_tokenize
train_data = []
for index ,row in train.iterrows():
    t = word_tokenize(row['text'])
    train_data.append((t,row['airline_sentiment']))    
train_data[0]
test_data = []
for index ,row in test.iterrows():
    t = word_tokenize(row['text'])
    test_data.append(t)    
# test_data[0]
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if(tag.startswith('J')):
        return wordnet.ADJ
    elif(tag.startswith('V')):
        return wordnet.VERB
    elif(tag.startswith('N')):
        return wordnet.NOUN
    elif(tag.startswith('R')):
        return wordnet.ADV
    else :
        return wordnet.NOUN
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
clean_data = [(clean_review(text),sentiment) for text,sentiment in train_data]
```

```python
clean_data[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
sentiments = [sentiment for text, sentiment in clean_data]
```

```python
text_documents = [" ".join(text) for text, sentiment in clean_data]
x_test = [" ".join(tweet) for tweet in test_data]
```

```python
text_documents
```

```python
count_vec = CountVectorizer(max_features = 2000,max_df = 0.8)
x_train_features = count_vec.fit_transform(text_documents)
x_test_features = count_vec.transform(x_test)
# x_train_features.todense()
```

```python
vectorizer = TfidfVectorizer(max_features=1000, max_df = 0.8)
x_train_features2 = vectorizer.fit_transform(text_documents)
x_test_features2 = vectorizer.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(C = 1000)
svc.fit(x_train_features2, sentiments)
```

```python
y_pred = svc.predict(x_test_features2)
y_pred
```

```python
svc.score(x_test_features2, y_pred)
```
